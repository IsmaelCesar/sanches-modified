```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
errors = [0.5, 0.1, 0.05, 0.01, 0.005, 0.001]
df_depth_mean_original = []
df_depth_mean_modified = []

for eps in errors:
    df_depth = pd.read_csv(f"circ_depth_error-{eps}.csv")
    df_depth_mean_original += [df_depth[["depth_original"]].to_numpy()]
    df_depth_mean_modified += [df_depth[["depth_modified"]].to_numpy()]

df_depth_mean_original = np.mean(df_depth_mean_original, axis=0)
df_depth_mean_modified = np.mean(df_depth_mean_modified, axis=0)

plt.title("Mean Circuit depth")
plt.plot(df_depth[["num_qubits"]], df_depth_mean_original, "--", marker=".",  label="original")
plt.plot(df_depth[["num_qubits"]], df_depth_mean_modified, "-.", marker="^", label="modified")
plt.ylabel("depth")
plt.xlabel("num qubits")
plt.legend(loc="best")
plt.show()
```

```python
errors = [0.5, 0.1, 0.05, 0.01, 0.005, 0.001]
fid_loss_mean_original = []
fid_loss_mean_modified = []

for eps in errors:
    df_fidloss = pd.read_csv(f"fid_loss_error-{eps}.csv")
    fid_loss_mean_original += [df_fidloss[["fidloss_original"]].to_numpy()]
    fid_loss_mean_modified += [df_fidloss[["fidloss_modified"]].to_numpy()]

fid_loss_mean_original = np.mean(fid_loss_mean_original, axis=0)
fid_loss_mean_modified = np.mean(fid_loss_mean_modified, axis=0)



plt.title("Mean Fidelity loss")
plt.plot(df_fidloss[["num_qubits"]], fid_loss_mean_original, "--",  marker=".",  label="original")
plt.plot(df_fidloss[["num_qubits"]], fid_loss_mean_modified, "-.",  marker="^", label="modified")
plt.ylabel("Fidelity loss")
plt.xlabel("num qubits")
plt.legend(loc="best")
plt.show()
```
